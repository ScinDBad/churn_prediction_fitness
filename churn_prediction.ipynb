### Pre-procesamiento

Los valores atípicos son abundantes en algunas características y en otras no. Según sea la distribución de las características se aplicará un determinado método de gestión de outliers.

```python
# Función de agregación personalizada similar a .describe() con métricas de outliers para todas las columnas
def count_outliers(group):
    # Inicializar un diccionario para almacenar las estadísticas calculadas
    summary_stats = {}

    # Iterar sobre cada columna en el grupo
    for col in group.columns:
        values = group[col]
               
        # Calcular outliers usando el rango intercuartílico (IQR)
        q1 = np.percentile(values, 25)
        q3 = np.percentile(values, 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = (values < lower_bound) | (values > upper_bound)
        num_outliers = outliers.sum()
        outlier_percentage = (num_outliers / values.count()) * 100 
        
        # Calcular percentil correspondiente a upper_bound
        upper_bound_percentile = stats.percentileofscore(values, upper_bound)

        # Almacenar las estadísticas en el diccionario
        summary_stats[col] = {
            'num_outliers': num_outliers,
            'outlier_%': outlier_percentage,
            'outlying_p()': upper_bound_percentile
        }

    # Convertir el diccionario en un DataFrame y retornar
    return pd.DataFrame(summary_stats)
```

#### Reemplazo de Outliers

```python
# Función para reemplazar outliers por el valor del percentil x de cada columna
def replace_outliers_with_percentile(df, features, percentile, p_upper_bound=None, obj=None, obj_value=None):
    # Copiar el DataFrame para no modificar el original
    df_cleaned = df.copy()
    mask = False  # Inicializar obj_mask con un valor por defecto
    
    # Iterar sobre cada columna característica del DataFrame
    for feature in features:
        # Calcular el percentil de la columna actual
        percentile_ = np.percentile(df_cleaned[feature], percentile)
        
        # Calcular los límites para identificar outliers
        q1 = np.percentile(df_cleaned[feature], 25)
        q3 = np.percentile(df_cleaned[feature], 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Reemplazar outliers por el valor del percentil        
        if p_upper_bound:
            upper_bound = np.percentile(df_cleaned[feature], p_upper_bound)
        
        # Condicion de segmento objetivo
        if (obj and obj_value):
            mask = (df_cleaned[obj]==obj_value)
        else:
            mask = True
        
        # Reemplazo de outliers según percentil, IQR y segmento objetivo
        df_cleaned[feature] = np.where(
            np.logical_and(
                mask, 
                (df_cleaned[feature] < lower_bound) | (df_cleaned[feature] > upper_bound)
            ),
            percentile_,
            df_cleaned[feature]
        )
    
    return df_cleaned
```

```python
churn_values = [0, 1]
gym_cleaned = gym

for churn_val in churn_values:
    # Reemplazo con P95
    gym_cleaned = replace_outliers_with_percentile(
        df=gym_cleaned, obj='churn', obj_value=churn_val,
        features=[
            'avg_additional_charges_total',
            'avg_class_frequency_total',
            'avg_class_frequency_current_month'
        ],
        percentile=95
    )  

    # Reemplazo parcial con P50
    gym_cleaned = replace_outliers_with_percentile(
        df=gym_cleaned, obj='churn', obj_value=churn_val,
        features=['lifetime'],
        percentile=50,
        p_upper_bound=97.5,
    )

    # Reemplazo con P50
    gym_cleaned = replace_outliers_with_percentile(
         df=gym_cleaned, obj='churn', obj_value=churn_val,
        features=['age'],
        percentile=50,
    )
```

```python
# estandariza los datos
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train) # estandariza el conjunto de entrenamiento 
X_test_st = scaler.transform(X_test) # estandariza el conjunto de prueba
```

### Entrenamiento y Predicción

```python
# Declarar modelos
seed = 42 # semilla para generación aleatoria
logr_model = LogisticRegression(random_state=seed)
rfc_model = RandomForestClassifier(random_state=seed)

def make_prediction(model, X_train, y_train, X_test, y_test):
    # Iniciar el temporizador para medir el tiempo de entrenamiento
    start_time = time.time()
    
    # Entrenar el modelo
    model.fit(X_train, y_train)
    
    # Calcular el tiempo de entrenamiento
    training_time = time.time() - start_time
    
    # Iniciar el temporizador para medir el tiempo de predicción
    start_time = time.time()
    
    # Hacer predicciones
    y_pred = model.predict(X_test)
    
    # Calcular el tiempo de predicción
    prediction_time = time.time() - start_time
    
    # Mostrar todas las métricas estudiadas para la predicción resultante
    print(model)
    print('Exactitud: {:.2f}'.format(accuracy_score(y_test, y_pred)))
    print('Precisión: {:.2f}'.format(precision_score(y_test, y_pred)))
    print('Recall: {:.2f}'.format(recall_score(y_test, y_pred)))
    
    print(f"\nTiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")    
```

```python
make_prediction(logr_model, X_train_st, y_train, X_test_st, y_test)
```

El modelo de regresión logística obtuvo mejores resultados y además es más rápido que el bosque aleatorio.

```python
# Función para graficar curvas de aprendizaje
def plot_learning_curve(estimator, X, y, model_name, score_reference):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.figure(figsize=(11.5, 6))
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="b")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="b", label="Cross-validation score")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.title(f"Learning Curve ({model_name})")
    plt.axhline(y=score_reference, color='g', linestyle='--', label="Desired Score (0.91)")
    plt.legend()
    plt.show()
```

```python
# Función para evaluar el modelo RandomForestClassifier
def evaluate_model_classifier(model, model_name, X_test, y_test):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw={'wspace': 0.5})  # Subplot de 1 fila y 2 columnas

    # Matriz de confusión
    disp_confusion = plot_confusion_matrix(model, X_test, y_test,
                                           cmap=plt.cm.Blues,
                                           normalize=None,
                                           ax=axes[0])
    disp_confusion.ax_.set_title(f'Confusion Matrix ({model_name})')

    # Curva ROC
    disp_roc = plot_roc_curve(model, X_test, y_test, ax=axes[1])
    disp_roc.ax_.set_title(f'ROC Curve ({model_name})')
    plt.show()
```

```python
# Evaluar modelo
start_time = time.time()

plot_learning_curve(logr_model, X_train_st, y_train, 'Logistic Regression', 0.91)
evaluate_model_classifier(logr_model, 'Logistic Regression', X_test_st, y_test)

print(f'ejecución de evaluación: {time.time() - start_time:.2f} segundos')
```

La curva de entrenamiento converge con la curva validación cruzada apartir de 1000 ejemplos de entrenamiento alcanzando un puntaje de 0.91 a los 2500 ejemplos.
```python
# función de trazado de barplot comparativo
def barplot(df):
    # Número de filas para el subplot, calculado según el número de características y columnas
    num_rows = int(np.ceil(len(df.columns) / 3))

    # Crear una figura y un conjunto de subplots
    fig, axes = plt.subplots(num_rows, 3, figsize=(11, num_rows * 2))

    # Aplanar el array de axes para iterar fácilmente
    axes = axes.flatten()
    
    # Bucle para dibujar cada histograma en su respectivo subplot
    for i, column in enumerate(df.columns):
        sns.barplot(data=df, x=df.index, y=column, ax=axes[i], color='blue', alpha=0.3)
        axes[i].set_title(f'Mean Rate - Cluster {column}')
        axes[i].set_ylim(0,1)
        axes[i].set_ylabel('rate')
        axes[i].set_xlabel('features')
        axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90, ha='center')
        if i in [0,1]:
            axes[i].set_xlabel('')
            axes[i].set_xticklabels([])  # Ocultar las etiquetas del eje x
        if i in [1,2,4]:
            axes[i].set_ylabel('')

    # Ocultar los axes vacíos si los hay
    for ax in axes[len(df.columns):]:
        ax.set_visible(False)

    # Mostrar la figura con todos los subplots
    #plt.tight_layout()
    plt.show()
```

Se procede a trazar los clústeres por par de características.

```python
# Definir función de trazado básico de clústeres por par-características
def show_clusters_on_plot(df, x_name, y_name, cluster_name, ax):
    sns.scatterplot(x=df[x_name], y=df[y_name], hue=df[cluster_name], palette='Paired', ax=ax)
    ax.set_title(f'Clustering scatter')

# Definir función para agrupar gráficas de clústers por par-características
def plot_clusters_subplots(df, feature_combinations, cluster_name, figsize=(13.5, 4.5)):
    num_plots = len(feature_combinations)
    num_cols = min(num_plots, 3)  # Limitar el número de columnas a 3 para visualizar hasta 3 gráficos por fila
    num_rows = (num_plots // num_cols) + (1 if num_plots % num_cols != 0 else 0)  # Calcular el número de filas necesarias
    
    fig, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
    axes = axes.flatten()  # Aplanar el array de axes para iterar fácilmente
    
    for i, (x_name, y_name) in enumerate(feature_combinations):
        if i < len(axes):  # Solo graficar si hay un subplot disponible
            show_clusters_on_plot(df, x_name, y_name, cluster_name, ax=axes[i])
        else:
            break  # Detener el bucle si no hay más subplots disponibles
    
    # Ocultar los axes vacíos si los hay
    for ax in axes[num_plots:]:
        ax.set_visible(False)
    
    plt.tight_layout()  # Ajustar el espacio entre subplots
    plt.show()
```

```python
# Obtener todas las combinaciones por pares
feature_combinations = list(combinations(selected_features_, 2))
gym_km = gym[list(selected_features_) + ['cluster_km']]
    
# trazar subplot
plot_clusters_subplots(gym_km, feature_combinations[:3], 'cluster_km')
```

#### Tasa de Cancelación

Se procede a calcular la proporción en la que cancelan los usuarios de cada cluster.

```python
# Calculo de tasa de abandano de usuarios del gimnasio
gym.groupby('cluster_km')['churn'].mean().round(4).sort_values(ascending=False) * 100
```

__Nota:__ Dado que los valores posibles de 'churn' son 1 y 0 el promedio de los valores de 1 representa particularmente también la tasa de abandono; esto porque el sumar los 1 equivale a contarlos. __Pero vamos a demostrarlo__, procederemos a realizar el conteo correspondiente.

```python
k_churn = gym.groupby('cluster_km')['churn'].agg({'count', 'sum'})
k_churn['churn_%'] = (k_churn['sum'] / k_churn['count'] * 100).round(2)
k_churn['mean'] = gym.groupby('cluster_km')['churn'].mean().round(4) * 100
k_churn.sort_values('mean', ascending=False)
```

Previamente en el trazado de histogramas en la distribución de clústeres, ya se había obtenido la media pero por el redondeo aplicado y sin multiplicar por 100 se perdió precisión. Ahora en el dataframe reciente se puede apreciar exactamente la tasa de cancelación de los clientes de cada clúster:

- El __0.12%__ de los usuarios del __cluster 0__ cancelaron sus membresías
- El __1.55%__ de los usuarios del __cluster 1__ cancelaron sus membresías
- El __0.71%__ de los usuarios del __cluster 2__ cancelaron sus membresías
- El __26.37%__ de los usuarios del __cluster 3__ cancelaron sus membresías
- El __99.79%__ de los usuarios del __cluster 4__ cancelaron sus membresías

Por lo tanto el quinto grupo es el más propenso a retirarse, por no decir en su totalidad; en segundo lugar el cuarto grupo (cluster 3) donde 26 de cada 100 usuarios abandonan el gym. En el resto de grupos por cada 100 usuarios menos de 2 de ellos cancelan sus membresías.
